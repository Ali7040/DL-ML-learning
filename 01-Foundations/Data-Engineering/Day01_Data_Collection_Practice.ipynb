# Part 1: Working with REST APIs 🌐

## Theory Refresher

**REST API** = Representational State Transfer Application Programming Interface
- A way to get data from servers over the internet
- Uses HTTP methods: GET, POST, PUT, DELETE
- Returns data usually in JSON format

**Common Status Codes:**
- 200: Success
- 404: Not Found
- 500: Server Error

## Exercise 1.1: Simple API Request

Let's fetch data from JSONPlaceholder (a fake REST API for testing).

```python
# Example: Fetch posts from JSONPlaceholder
url = 'https://jsonplaceholder.typicode.com/posts'

# Make GET request
response = requests.get(url)

# Check status
print(f"Status Code: {response.status_code}")
print(f"Content Type: {response.headers['Content-Type']}")

# Parse JSON
data = response.json()

# Display first 3 posts
print(f"\n📊 Total posts: {len(data)}")
print("\nFirst post:")
print(json.dumps(data[0], indent=2))
```

## Exercise 1.4: GitHub API - Real-World Example

**TASK:** Fetch information about a GitHub repository.

```python
# Example: Get Python repository info
repo_url = 'https://api.github.com/repos/python/cpython'

response = requests.get(repo_url)

if response.status_code == 200:
    repo_data = response.json()
    
    # Extract interesting information
    info = {
        'name': repo_data['name'],
        'full_name': repo_data['full_name'],
        'description': repo_data['description'],
        'stars': repo_data['stargazers_count'],
        'forks': repo_data['forks_count'],
        'language': repo_data['language'],
        'created_at': repo_data['created_at'],
        'updated_at': repo_data['updated_at']
    }
    
    for key, value in info.items():
        print(f"{key}: {value}")
else:
    print(f"Error: {response.status_code}")
```

# Part 2: Reading Files 📁

## Theory: File Formats

Common data formats:
- **CSV**: Comma-Separated Values (most common)
- **JSON**: JavaScript Object Notation (web APIs)
- **Excel**: .xlsx, .xls (business data)
- **Parquet**: Columnar format (big data)
- **TSV**: Tab-Separated Values

## Exercise 2.1: Create Sample Data

First, let's create sample datasets to practice with.

```python
# Create sample data
import numpy as np

# Sample dataset: Student grades
students_data = {
    'student_id': range(1, 101),
    'name': [f'Student_{i}' for i in range(1, 101)],
    'math_score': np.random.randint(50, 100, 100),
    'science_score': np.random.randint(45, 100, 100),
    'english_score': np.random.randint(55, 100, 100),
    'grade': np.random.choice(['A', 'B', 'C', 'D'], 100)
}

df_students = pd.DataFrame(students_data)

print("Sample Student Data:")
print(df_students.head(10))
print(f"\nShape: {df_students.shape}")
```

## Exercise 2.2: Save to Different Formats

```python
# Save to CSV
df_students.to_csv('students.csv', index=False)
print("✅ Saved to CSV")

# Save to JSON
df_students.to_json('students.json', orient='records', indent=2)
print("✅ Saved to JSON")

# Save to Excel
df_students.to_excel('students.xlsx', index=False)
print("✅ Saved to Excel")

print("\n📁 All files created successfully!")
```

## Exercise 2.4: Reading CSV with Options

```python
# Create a messy CSV
messy_data = """# This is a comment
# Student data from 2026
ID;Name;Score;Date
1;Alice;95;2026-03-01
2;Bob;87;2026-03-02
3;Charlie;92;2026-03-03
skip_this_row;invalid;data;here
4;David;88;2026-03-04
"""

# Save to file
with open('messy_students.csv', 'w') as f:
    f.write(messy_data)

# Read with options
df_messy = pd.read_csv(
    'messy_students.csv',
    sep=';',  # Delimiter
    comment='#',  # Skip comment lines
    skiprows=lambda x: 'skip' in str(x),  # Skip specific rows
    parse_dates=['Date']  # Parse date column
)

print("Cleaned data:")
print(df_messy)
print(f"\nData types:\n{df_messy.dtypes}")
```

# Part 4: Working with Databases 🗄️

## Exercise 4.1: Create SQLite Database

```python
# Create SQLite database
conn = sqlite3.connect('student_db.sqlite')

# Write DataFrame to database
df_students.to_sql('students', conn, if_exists='replace', index=False)

print("✅ Database created and data inserted!")

# Verify
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(f"\nTables in database: {cursor.fetchall()}")
```

## Exercise 4.2: Query Database

```python
# Read data using SQL query
query = "SELECT * FROM students WHERE math_score > 90 LIMIT 10"
df_query = pd.read_sql_query(query, conn)

print("Students with Math score > 90:")
print(df_query)
```
